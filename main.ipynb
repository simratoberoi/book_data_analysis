```python
import pandas as pd
df = pd.read_csv("bestsellers.csv")
print(df.head())
print(df.tail())
print(df.shape)
print(df.columns)
print(df.describe())
df.drop_duplicates(inplace=True)
df.rename(columns={"Name": "Title", "Year": "Publication Year", "User Rating": "Rating"}, inplace=True)
df["Price"] = df["Price"].astype(float)
top_authors=df["Author"].value_counts()
print(top_authors)
top_authors.head(10).to_csv("top_authors.csv")
avg_genre_rating=df.groupby("Genre")["Rating"].mean()
print(avg_genre_rating)
avg_genre_rating.to_csv("avg_genre_rating.csv")
max_price=df["Price"].max()
print("the max price is: ", max_price)
year_rating=df.groupby("Publication Year")["Rating"].mean()
print(year_rating)
year_rating.to_csv("yearwise_rating.csv")
authors=df.groupby("Author")["Author"].value_counts()
print(authors)
authors.to_csv("Authors.csv")
```
